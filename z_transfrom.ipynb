```python
def Z_Transform(scales, causal, z):
    X_z = np.zeros_like(z, dtype=complex)
    
    for k in range(len(scales)): 
        for r in range(len(z[0])):   
            for c in range(len(z[1])):  
                if causal[k]:
                    X_z[r, c] += (scales[k]**r) * z[r, c]**-c
                else:
                    X_z[r, c] += (scales[k]**-r) * z[r, c]**c
    
    roc = []
    for i in range(len(scales)):  # Generate ROC based on causality
        if causal[i]:
            roc.append(f"|Z| > {scales[i]}")
        else:
            roc.append(f"|Z| < {scales[i]}")
    
    roc = " and ".join(roc)
    
    return X_z, f"ROC: {roc}"
```

```python
scales = [np.exp(-0.2)]
causal = [True]

r = np.logspace(0, 2, 10)
theta = np.linspace(0, 2 * np.pi, 10)
z = r[:, None] * np.exp(1j * theta[None, :])

roc = "Entire Z-Plane"
X_Z, roc = Z_Transform(scales, causal, z)

# Flatten the array for easy output
X_Z = X_Z.flatten()

print("Z-Transformation values:")
for i, z_val in enumerate(X_Z):
    print(f"Z[{i}]: {z_val}")

print(roc)
```

```python

magnitudes = np.abs(X_Z) 

phases = np.angle(X_Z)

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

axs[0].set_title("Magnitude of Z-Transform Inside Unit Circle")

circle = plt.Circle((0, 0), 1, color='k', linestyle='-', fill=False, label="Unit Circle")
axs[0].add_artist(circle)

for i in range(len(magnitudes)):
    axs[0].plot(np.real(X_Z[i]), np.imag(X_Z[i]), 'bo', label="Magnitude" if i == 0 else "")  

    axs[0].plot([0, np.real(X_Z[i])], [0, np.imag(X_Z[i])], 'b-') 


axs[0].set_xlim(-1, 1)
axs[0].set_ylim(-1, 1)  
axs[0].set_aspect('equal')  
axs[0].legend()
axs[0].grid(True)

axs[1].set_title("Phase of Z-Transform on Unit Circle")

circle = plt.Circle((0, 0), 1, color='k', linestyle='-', fill=False, label="Unit Circle")
axs[1].add_artist(circle)

for i in range(len(phases)):
    axs[1].plot(np.cos(phases[i]), np.sin(phases[i]), 'ro', label="Phase" if i == 0 else "") 
    axs[1].plot([0, np.cos(phases[i])], [0, np.sin(phases[i])], 'r-')  

axs[1].set_xlim(-1, 1)
axs[1].set_ylim(-1, 1) 
axs[1].set_aspect('equal')  
axs[1].legend()
axs[1].grid(True)

plt.show()
```

```python
def Z_Transform(omega, causal, z):
    X_z = np.zeros_like(z, dtype=complex)
    
    for k in range(len(causal)):
        e_j_omega = np.exp(1j * omega[k]) 
        e_j__omega = np.exp(-1j * omega[k]) 
        for r in range(z.shape[0]):
            for c in range(z.shape[1]):
                if causal[k]:
                    X_z[r, c] += (1 / 2j) * (((e_j_omega * z[r, c]**-1))**c)
                    X_z[r, c] -= (1 / 2j) * (((e_j__omega * z[r, c]**-1))**c)
                else:
                    X_z[r, c] -= (-1 / 2j) * (((e_j_omega**-1 * z[r, c]))**c)
                    X_z[r, c] += (-1 / 2j) * (((e_j__omega**-1 * z[r, c]))**c)

    roc = "Entire Z-Plane"
    if causal:
        roc = "|z| > 1"

    return X_z, f"ROC: {roc}"
```

```python
omega = [np.pi / 4]
causal = [True]

r = np.logspace(0, 2, 10)  
theta = np.linspace(0, 2 * np.pi, 10) 
z = r[:, None] * np.exp(1j * theta[None, :])

X_Z, roc = Z_Transform(omega, causal, z)

X_Z = X_Z.flatten()
z = z.flatten()
print("Z-Transformation values:")
for i, z in enumerate(X_Z):
    print(f"Z[{i}]: {z}")

print(roc)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example Z-transformed values
X_Z = np.array([0.1 + 0.3j, 0.2 + 0.1j, 0.4 + 0.2j, 0.3 + 0.4j, 0.1 + 0.5j])

# Magnitudes and Phases
magnitudes = np.abs(X_Z)  # Magnitudes
phases = np.angle(X_Z)    # Phases

# Create the figure
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# 🔹 Magnitude Plot (inside the unit circle)
axs[0].set_title("Magnitude of Z-Transform Inside Unit Circle")

# Draw the full unit circle (border only, no filling)
circle = plt.Circle((0, 0), 1, color='k', linestyle='-', fill=False, label="Unit Circle")
axs[0].add_artist(circle)

# Plotting magnitudes (scaled to fit inside the unit circle)
for i in range(len(magnitudes)):
    # Plot each magnitude point inside the circle
    axs[0].plot(np.real(X_Z[i]), np.imag(X_Z[i]), 'bo', label="Magnitude" if i == 0 else "")  

# Limit to unit circle boundary
axs[0].set_xlim(-1, 1)
axs[0].set_ylim(-1, 1)  # Both X and Y axes range from -1 to 1
axs[0].set_aspect('equal')  # Ensure a proper circle aspect ratio
axs[0].legend()
axs[0].grid(True)

# 🔹 Phase Plot (on the unit circle)
axs[1].set_title("Phase of Z-Transform on Unit Circle")

# Draw the full unit circle (border only, no filling)
circle = plt.Circle((0, 0), 1, color='k', linestyle='-', fill=False, label="Unit Circle")
axs[1].add_artist(circle)

# Plotting the phase points around the unit circle
for i in range(len(phases)):
    axs[1].plot(np.cos(phases[i]), np.sin(phases[i]), 'ro', label="Phase" if i == 0 else "")  # Phase points

# Limit to unit circle boundary
axs[1].set_xlim(-1, 1)
axs[1].set_ylim(-1, 1)  # Both X and Y axes range from -1 to 1
axs[1].set_aspect('equal')  # Ensure a proper circle aspect ratio
axs[1].legend()
axs[1].grid(True)

plt.show()
```

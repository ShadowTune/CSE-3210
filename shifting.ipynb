```python
import numpy as np
import matplotlib.pyplot as plt

# Time axis
t = np.linspace(-1, 1, 1000)  # Time vector from -1 to 1 second

# Original signal: Gaussian pulse
original_signal = np.exp(-((t - 0)**2) / (2 * 0.1**2))  # Centered at t=0, width=0.1

# Time-shifted signals
shift_1 = 0.3  # Positive shift (delay)
shift_2 = -0.3  # Negative shift (advance)
signal_shifted_1 = np.exp(-((t - shift_1)**2) / (2 * 0.1**2))
signal_shifted_2 = np.exp(-((t - shift_2)**2) / (2 * 0.1**2))

# Plot signals
plt.figure(figsize=(12, 6))

# Original signal
plt.plot(t, original_signal, label="Original Signal", color="blue")

# Time-shifted signals
plt.plot(t, signal_shifted_1, label=f"Delayed Signal (+{shift_1}s)", linestyle="--", color="green")
plt.plot(t, signal_shifted_2, label=f"Advanced Signal ({shift_2}s)", linestyle=":", color="red")

# Add labels, legend, and grid
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Visualization of Time Shifting with a Gaussian Pulse")
plt.axhline(0, color="black", linewidth=2, linestyle="--")
plt.axvline(0, color="black", linewidth=2, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.signal import square

# Original signal: Square wave
original_signal = square(2 * np.pi * 2 * t)  # 2 Hz square wave

# Time-shifted signals
signal_shifted_1 = square(2 * np.pi * 2 * (t - shift_1))  # Delayed
signal_shifted_2 = square(2 * np.pi * 2 * (t - shift_2))  # Advanced

# Plot signals
plt.figure(figsize=(12, 6))

# Original signal
plt.plot(t, original_signal, label="Original Signal", color="blue")

# Time-shifted signals
plt.plot(t, signal_shifted_1, label=f"Delayed Signal (+{shift_1}s)", linestyle="--", color="green")
plt.plot(t, signal_shifted_2, label=f"Advanced Signal ({shift_2}s)", linestyle=":", color="red")

# Add labels, legend, and grid
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Visualization of Time Shifting with a Square Wave")
plt.axhline(0, color="black", linewidth=2, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f = 5  # Frequency (Hz)
t = np.linspace(0, 1, 1000)  # Time vector (1 second, 1000 samples)

# Original signal
x = np.sin(2 * np.pi * f * t)

# Phase-shifted signals
phi_90 = np.pi / 2  # 90 degrees
phi_180 = np.pi     # 180 degrees
phi_270 = 3 * np.pi / 2  # 270 degrees

x_90 = np.sin(2 * np.pi * f * t + phi_90)
x_180 = np.sin(2 * np.pi * f * t + phi_180)
x_270 = np.sin(2 * np.pi * f * t + phi_270)

# Plot signals
plt.figure(figsize=(12, 6))
plt.plot(t, x, label="Original Signal (0)", color="blue")
plt.plot(t, x_90, label="90 Phase Shift", color="green", linestyle="--")
plt.plot(t, x_180, label="180 Phase Shift", color="red", linestyle=":")
plt.plot(t, x_270, label="270 Phase Shift", color="purple", linestyle="-.")
plt.axhline(0, color='black', linewidth=2, linestyle='--')
plt.axvline(0, color='black', linewidth=2, linestyle='--')
plt.title("Phase Shifting of a Sine Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Original signal: Sine wave
t_original = np.linspace(0, 1, 10)  # 10 samples
signal_original = np.sin(2 * np.pi * t_original)

# Upscaling: Interpolation
t_upsampled = np.linspace(0, 1, 100)  # 100 samples
interpolator = interp1d(t_original, signal_original, kind='cubic')  # Cubic interpolation
signal_upsampled = interpolator(t_upsampled)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(t_original, signal_original, 'o', label="Original Samples")
plt.plot(t_upsampled, signal_upsampled, label="Upsampled Signal", linestyle="--")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Signal Upscaling using Cubic Interpolation")
plt.legend()
plt.grid()
plt.show()
```

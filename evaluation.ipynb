```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
def convolution(a, b):
    n = len(a)
    m = len(b)
    result = [0] * (n + m - 1)

    for i in range(n):
        for j in range(m):
            result[i + j] += a[i] * b[j]
    
    return result

a = int(input(""))
a2 = int(input(""))
alimit = int(input(""))
x = np.linspace(a, a2, alimit)

b = int(input(""))
b2 = int(input(""))
blimit = int(input(""))
y = np.linspace(b, b2, blimit)

print(x)
print(y)
result = convolution(x, y)
print(result)
print(np.convolve(x, y, mode="full"))
```

```python
# Plot both convolutions in a single figure
plt.figure(figsize=(14, 7))
plt.suptitle('Comparison of Convolution Results', fontsize=20)
style.use('dark_background')

# Plot the numpy convolution
plt.plot(np.convolve(x, y, mode="full"), 'r-', linewidth=3, label='Convolution by np.convolve')

# Plot the Wezney method convolution
plt.plot(result, 'b--', linewidth=3, label='Convolution by Wezney method')

# Set axis labels, legend, and grid
plt.xlabel('Sample', fontsize=15)
plt.ylabel('Value', fontsize=15)
plt.legend(fontsize=15)
plt.grid(True)

# Show the combined plot
plt.show()
```

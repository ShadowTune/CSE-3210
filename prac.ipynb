```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import style
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]
plt.plot(x, y, label = "Y = 2X", linestyle='--',
         color = "black", marker='.', markersize=10)
plt.xlabel("Time")
plt.ylabel("Temp")
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2, x2**2, 'r', label='Y = X^2')
plt.legend()
plt.show()
print(x2)
```

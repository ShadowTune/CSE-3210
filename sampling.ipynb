```python
import numpy as np
import matplotlib.pyplot as plt

# Original signal: Y = sin(x)
def original_signal(x):
    return np.sin(x)

# Linear interpolation function
def linear_interpolation(x_samples, y_samples, x_interp):
    y_interp = []
    for x in x_interp:
        # Find the interval [x0, x1] where x lies
        for i in range(len(x_samples) - 1):
            if x_samples[i] <= x <= x_samples[i + 1]:
                x0, y0 = x_samples[i], y_samples[i]
                x1, y1 = x_samples[i + 1], y_samples[i + 1]
                # Linear interpolation formula
                y = y0 + (y1 - y0) / (x1 - x0) * (x - x0)
                y_interp.append(y)
                break
    return np.array(y_interp)

# Generate data for the original signal
x_original = np.linspace(0, 2 * np.pi, 500)  # Continuous range
y_original = original_signal(x_original)

# Sample points
x_samples = np.linspace(0, 2 * np.pi, 10)  # Discrete sample points
y_samples = original_signal(x_samples)

# Interpolation
x_interp = np.linspace(0, 2 * np.pi, 500)  # High-resolution points for interpolation
y_interp = linear_interpolation(x_samples, y_samples, x_interp)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_original, y_original, label='Original Signal (Y = sin(x))', color='blue', linestyle='--')
plt.scatter(x_samples, y_samples, label='Sample Points', color='red', zorder=5)
plt.plot(x_interp, y_interp, label='Linear Interpolation', color='green')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Signal Sampling and Linear Interpolation for Y = sin(x)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Linear interpolation function
def linear_interpolation(x_samples, y_samples, x_interp):
    y_interp = []
    for x in x_interp:
        # Find the interval [x0, x1] where x lies
        for i in range(len(x_samples) - 1):
            if x_samples[i] <= x <= x_samples[i + 1]:
                x0, y0 = x_samples[i], y_samples[i]
                x1, y1 = x_samples[i + 1], y_samples[i + 1]
                # Linear interpolation formula
                y = y0 + (y1 - y0) / (x1 - x0) * (x - x0)
                y_interp.append(y)
                break
    return np.array(y_interp)

# Function to quantize the signal
def quantize_signal(signal, levels, min_val, max_val):
    """
    Quantizes the input signal into discrete levels.

    Parameters:
    - signal: The input signal (array).
    - levels: Number of quantization levels.
    - min_val: Minimum value of the range.
    - max_val: Maximum value of the range.

    Returns:
    - Quantized signal.
    """
    step_size = (max_val - min_val) / levels  # Calculate quantization step size
    quantized_signal = np.round((signal - min_val) / step_size) * step_size + min_val
    return np.clip(quantized_signal, min_val, max_val)  # Ensure values stay within range

# Generate the original signal: Y = sin(x)
x = np.linspace(0, 360, 500)  # X values (degrees)
y_original = np.sin(np.radians(x))  # Original signal in radians

# Sample points
x_samples = np.linspace(0, 360, 50)  # Discrete sample points in degrees
y_samples = np.sin(np.radians(x_samples))  # Corrected


# Interpolation
x_interp = np.linspace(0, 360, 500)  # High-resolution points for interpolation in degrees
y_interp = linear_interpolation(x_samples, y_samples, x_interp)

# Quantization parameters
num_levels = 8  # Number of quantization levels
y_min = -1  # Minimum value of the range (sin(x) range)
y_max = 1  # Maximum value of the range

# Perform quantization
y_quantized = quantize_signal(y_original, num_levels, y_min, y_max)

# Plot the results
plt.figure(figsize=(12, 6))

# Original signal
plt.plot(x, y_original, label="Original Signal (Y = sin(X))", color="blue", linestyle="--")

#Sampled signal
plt.scatter(x_samples, y_samples, label='Sample Points', color='red', zorder=5)

plt.plot(x_interp, y_interp, label='Linear Interpolation', color='green')

# Quantized signal
plt.step(x, y_quantized, label=f"Quantized Signal ({num_levels} Levels)", color="red", where="mid")

# Add labels and legend
plt.xlabel("X (Degrees)")
plt.ylabel("Y")
plt.title("Signal Quantization: Y = sin(x)")
plt.grid()
plt.legend()
plt.xticks(np.arange(0, 361, 45))  # X-axis ticks in degrees
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Time axis
t = np.linspace(0, 1, 1000)  # 1 second, 1000 samples

# Original signal
f = 5000  # Frequency (Hz)
x = np.sin(2 * np.pi * f * t)

# Time-shifted signal
t_shift = 2  # Time shift (seconds)
x_time_shifted = np.sin(2 * np.pi * f * (t - t_shift))

# Frequency-shifted signal
f_shift = 10  # Frequency shift (Hz)
x_freq_shifted = x * np.cos(2 * np.pi * f_shift * t)

# Plot signals
plt.figure(figsize=(12, 8))

# Original signal
plt.subplot(3, 1, 1)
plt.plot(t, x, label="Original Signal")
plt.title("Original Signal")
plt.grid()
plt.legend()

# Time-shifted signal
plt.subplot(3, 1, 2)
plt.plot(t, x_time_shifted, label=f"Time-Shifted Signal ({t_shift}s)")
plt.title("Time-Shifted Signal")
plt.grid()
plt.legend()

# Frequency-shifted signal
plt.subplot(3, 1, 3)
plt.plot(t, x_freq_shifted, label=f"Frequency-Shifted Signal ({f_shift} Hz)")
plt.title("Frequency-Shifted Signal")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
